```python
import pandas as pd
import numpy as np

# 1. CARGA DEL ARCHIVO COMPLETO
df_bank = pd.read_csv(ruta_bank, sep=',', header='infer', index_col=None)

# 2. LIMPIEZA DE EDAD, CATEGORÍAS Y ÍNDICES
df_bank = df_bank.dropna(subset=['housing', 'loan'])

mediana_edad = df_bank['age'].median()
df_bank['age'] = df_bank['age'].fillna(mediana_edad).astype(int)

columnas_texto = ['job', 'marital', 'education', 'default']
for col in columnas_texto:
    df_bank[col] = df_bank[col].fillna('unknown')

if 'Unnamed: 0' in df_bank.columns:
    df_bank = df_bank.drop(columns=['Unnamed: 0'])

# 3. LIMPIEZA DE COLUMNAS NUMÉRICAS CON ENTRADAS EXTRAÑAS
for col in ['cons.price.idx', 'euribor3m']:
    if df_bank[col].dtype == 'object':
        df_bank[col] = df_bank[col].str.replace(',', '.', regex=False)
    df_bank[col] = pd.to_numeric(df_bank[col], errors='coerce')
    df_bank[col] = df_bank[col].fillna(df_bank[col].median())

# 4. CORRECCIÓN DE LA FECHA (Cambiando el dataframe in-place de forma segura)
df_bank['date'] = pd.to_datetime(df_bank['date'], errors='coerce', format='mixed')

if not df_bank['date'].dropna().empty:
    fecha_moda = df_bank['date'].dropna().mode().values[0]
    df_bank['date'] = df_bank['date'].fillna(fecha_moda)
else:
    df_bank['date'] = df_bank['date'].fillna(pd.Timestamp('2026-06-18'))

# 5. COMPROBACIÓN FINAL
df_bank.info()
```
